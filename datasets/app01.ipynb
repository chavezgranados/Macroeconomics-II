Fuente: BCRP
*   X: exportaciones FOB (en millones de USD)
*   E: tipo de cambio intercambiario promedio

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('dataset01.xlsx')
df.head()
```

```python
# Grafica
plt.figure(figsize=(9, 6))  # Adjust figure size as needed
plt.scatter(df['E'], df['X'])
plt.xlabel('E')
plt.ylabel('X')
plt.title('Dispersion of X and E')
plt.show()
```
